```python
# for numbers
pf['Sales']=pd.to_numeric(pf['Sales'],errors='coerce')
# for Date Time
pf['Order Date']=pd.to_datetime(pf['Order Date'],errors='NaN')
```

```python
# Data Visualtion 
category_sales=pf.groupby('Category')['Sales'].sum()
category_sales
```

```python
plt.figure(figsize=(10,15))
category_sales.plot(kind='line',marker='o',color='black')
plt.xlabel("Category of Super Store")
plt.ylabel('total sales')
plt.title('Total sales by each category')
plt.grid(True,linestyle="--",alpha=1)
plt.show()
```

```python
category_sales=pf.groupby('Region')['Sales'].sum()
category_sales
```

```python
plt.figure(figsize=(6,4))
category_sales.plot(kind='bar',color=['red','green','violet','orange'])
plt.xlabel("Region of Super Store")
plt.ylabel("Total Sales")
plt.title("Total Sales from each region")
plt.grid(True,linestyle="--",alpha=0.6)
plt.show()
```
